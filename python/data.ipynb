```python
def preprocess_data(train_df, test_df):
    """Veriyi AI modeli i√ßin hazƒ±rla"""
    
    print("üîÑ Veri √∂n i≈üleme ba≈ülƒ±yor...")
    
    # Kategorik deƒüi≈ükenleri encode et
    categorical_cols = ['protocol_type', 'service', 'flag']
    label_encoders = {}
    
    for col in categorical_cols:
        le = LabelEncoder()
        # Train ve test i√ßin aynƒ± encoder'ƒ± kullan
        train_df[col] = le.fit_transform(train_df[col].astype(str))
        test_df[col] = le.transform(test_df[col].astype(str))
        label_encoders[col] = le
    
    # Sayƒ±sal √∂zellikleri se√ß (zorluk seviyesi ve hedef deƒüi≈ükenleri hari√ß)
    numerical_cols = [col for col in train_df.columns 
                     if col not in ['attack_type', 'difficulty_level', 'binary_label']]
    
    # Normalizasyon
    scaler = StandardScaler()
    train_df[numerical_cols] = scaler.fit_transform(train_df[numerical_cols])
    test_df[numerical_cols] = scaler.transform(test_df[numerical_cols])
    
    print("‚úÖ Veri √∂n i≈üleme tamamlandƒ±!")
    print(f"üìä Eƒüitim verisi: {train_df[numerical_cols].shape}")
    print(f"üìä Test verisi: {test_df[numerical_cols].shape}")
    
    return train_df, test_df, numerical_cols, label_encoders, scaler

# √ñn i≈üleme uygula
train_df, test_df, feature_cols, encoders, scaler = preprocess_data(train_df, test_df)
```

```python
def train_first_model(train_df, test_df, feature_cols):
    """ƒ∞lk AI modelini eƒüit"""
    
    print("ü§ñ AI Modeli eƒüitiliyor...")
    
    # √ñzellikler ve hedefler
    X_train = train_df[feature_cols]
    y_train = train_df['binary_label']
    
    X_test = test_df[feature_cols]
    y_test = test_df['binary_label']
    
    # Random Forest modeli
    model = RandomForestClassifier(
        n_estimators=100,
        max_depth=20,
        random_state=42,
        n_jobs=-1  # T√ºm CPU'larƒ± kullan
    )
    
    # Model eƒüitimi
    model.fit(X_train, y_train)
    
    # Tahminler
    y_pred = model.predict(X_test)
    
    # Sonu√ßlar
    accuracy = accuracy_score(y_test, y_pred)
    
    print(f"üéØ Model Doƒüruluƒüu: {accuracy:.4f} ({accuracy:.2%})")
    print("\nüìä Detaylƒ± Rapor:")
    print(classification_report(y_test, y_pred, target_names=['Normal', 'Saldƒ±rƒ±']))
    
    # Confusion Matrix
    plt.figure(figsize=(8, 6))
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Normal', 'Saldƒ±rƒ±'], 
                yticklabels=['Normal', 'Saldƒ±rƒ±'])
    plt.title('Confusion Matrix - Saldƒ±rƒ± Tespiti')
    plt.ylabel('Ger√ßek Deƒüer')
    plt.xlabel('Tahmin')
    plt.show()
    
    return model, accuracy

# Modeli eƒüit
model, accuracy = train_first_model(train_df, test_df, feature_cols)
```

```python
def plot_feature_importance(model, feature_cols):
    """En √∂nemli √∂zellikleri g√∂rselle≈ütir"""
    
    feature_imp = pd.DataFrame({
        'feature': feature_cols,
        'importance': model.feature_importances_
    }).sort_values('importance', ascending=False).head(15)
    
    plt.figure(figsize=(10, 8))
    sns.barplot(data=feature_imp, x='importance', y='feature')
    plt.title('Aƒü Saldƒ±rƒ± Tespiti i√ßin En √ñnemli 15 √ñzellik')
    plt.xlabel('√ñnemlilik Skoru')
    plt.tight_layout()
    plt.show()
    
    print("üèÜ En √ñnemli 10 √ñzellik:")
    for i, row in feature_imp.head(10).iterrows():
        print(f"  {row['feature']}: {row['importance']:.4f}")
    
    return feature_imp

# √ñzellik √∂nemliliƒüini g√∂ster
important_features = plot_feature_importance(model, feature_cols)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

def optimize_random_forest(train_df, test_df, feature_cols):
    """Random Forest hyperparameter tuning"""
    
    print("üîÑ Hyperparameter Optimizasyonu Ba≈ülƒ±yor...")
    
    X_train = train_df[feature_cols]
    y_train = train_df['binary_label']
    X_test = test_df[feature_cols]
    y_test = test_df['binary_label']
    
    # Parametre grid'i
    param_grid = {
        'n_estimators': [50, 100, 200],
        'max_depth': [10, 20, None],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    }
    
    # Grid Search
    rf = RandomForestClassifier(random_state=42, n_jobs=-1)
    grid_search = GridSearchCV(
        estimator=rf,
        param_grid=param_grid,
        cv=3,
        scoring='accuracy',
        n_jobs=-1,
        verbose=1
    )
    
    print("üîç En iyi parametreler aranƒ±yor...")
    grid_search.fit(X_train, y_train)
    
    # En iyi model
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    
    print(f"‚úÖ En iyi parametreler: {grid_search.best_params_}")
    print(f"üéØ Optimize Edilmi≈ü Doƒüruluk: {accuracy:.4f} ({accuracy:.2%})")
    
    return best_model, grid_search.best_params_

# Hyperparameter tuning uygula
print("1. ADIM: Hyperparameter Tuning")
optimized_model, best_params = optimize_random_forest(train_df, test_df, feature_cols)
```

```python
# 5. ADIM: Neural Network Modeli
print("üöÄ 5. ADIM: Neural Network Modeli ba≈ülatƒ±lƒ±yor...")

def create_neural_network(input_dim):
    """Basit bir neural network modeli olu≈ütur"""
    
    model = Sequential([
        Dense(128, activation='relu', input_shape=(input_dim,)),
        Dropout(0.3),
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])
    
    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    
    return model

def train_neural_network(train_df, test_df, feature_cols):
    """Neural network modeli eƒüit"""
    
    print("5. ADIM: Neural Network Modeli")
    
    X_train = train_df[feature_cols].values
    X_test = test_df[feature_cols].values
    y_train = train_df['binary_label'].values
    y_test = test_df['binary_label'].values
    
    # Modeli olu≈ütur
    nn_model = create_neural_network(X_train.shape[1])
    
    print("üß† Neural Network Modeli Summary:")
    nn_model.summary()
    
    print("ü§ñ Model eƒüitimi ba≈ülƒ±yor... (Bu biraz zaman alabilir)")
    
    # Modeli eƒüit
    history = nn_model.fit(
        X_train, y_train,
        epochs=50,
        batch_size=32,
        validation_split=0.2,
        verbose=1
    )
    
    # Deƒüerlendirme
    test_loss, test_accuracy = nn_model.evaluate(X_test, y_test, verbose=0)
    print(f"üéØ Neural Network Doƒüruluk: {test_accuracy:.4f} ({test_accuracy:.2%})")
    
    # Eƒüitim ge√ßmi≈üini g√∂rselle≈ütir
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Eƒüitim Doƒüruluƒüu')
    plt.plot(history.history['val_accuracy'], label='Doƒürulama Doƒüruluƒüu')
    plt.title('Model Doƒüruluƒüu')
    plt.xlabel('Epoch')
    plt.ylabel('Doƒüruluk')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Eƒüitim Kaybƒ±')
    plt.plot(history.history['val_loss'], label='Doƒürulama Kaybƒ±')
    plt.title('Model Kaybƒ±')
    plt.xlabel('Epoch')
    plt.ylabel('Kayƒ±p')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    return nn_model, test_accuracy, history

# Neural network modelini eƒüit
print("üß† Neural Network modeli eƒüitiliyor...")
nn_model, nn_accuracy, nn_history = train_neural_network(train_df, test_df, feature_cols)
print("‚úÖ 5. ADIM tamamlandƒ±!")
```

```python
# √ñnce gerekli mod√ºlleri import edelim
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
import tensorflow as tf

print("‚úÖ TensorFlow/Keras mod√ºlleri import edildi")

# ≈ûimdi 5. ADIM: Neural Network Modeli
print("üöÄ 5. ADIM: Neural Network Modeli ba≈ülatƒ±lƒ±yor...")

def create_neural_network(input_dim):
    """Basit bir neural network modeli olu≈ütur"""
    
    model = Sequential([
        Dense(128, activation='relu', input_shape=(input_dim,)),
        Dropout(0.3),
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])
    
    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    
    return model

def train_neural_network(train_df, test_df, feature_cols):
    """Neural network modeli eƒüit"""
    
    print("5. ADIM: Neural Network Modeli")
    
    X_train = train_df[feature_cols].values
    X_test = test_df[feature_cols].values
    y_train = train_df['binary_label'].values
    y_test = test_df['binary_label'].values
    
    # Modeli olu≈ütur
    nn_model = create_neural_network(X_train.shape[1])
    
    print("üß† Neural Network Modeli Summary:")
    nn_model.summary()
    
    print("ü§ñ Model eƒüitimi ba≈ülƒ±yor... (Bu biraz zaman alabilir)")
    
    # Modeli eƒüit (epoch'u 30'a d√º≈ü√ºrd√ºm daha hƒ±zlƒ± olmasƒ± i√ßin)
    history = nn_model.fit(
        X_train, y_train,
        epochs=30,  # 50'den 30'a d√º≈ü√ºrd√ºm
        batch_size=32,
        validation_split=0.2,
        verbose=1
    )
    
    # Deƒüerlendirme
    test_loss, test_accuracy = nn_model.evaluate(X_test, y_test, verbose=0)
    print(f"üéØ Neural Network Doƒüruluk: {test_accuracy:.4f} ({test_accuracy:.2%})")
    
    # Eƒüitim ge√ßmi≈üini g√∂rselle≈ütir
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Eƒüitim Doƒüruluƒüu')
    plt.plot(history.history['val_accuracy'], label='Doƒürulama Doƒüruluƒüu')
    plt.title('Model Doƒüruluƒüu')
    plt.xlabel('Epoch')
    plt.ylabel('Doƒüruluk')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Eƒüitim Kaybƒ±')
    plt.plot(history.history['val_loss'], label='Doƒürulama Kaybƒ±')
    plt.title('Model Kaybƒ±')
    plt.xlabel('Epoch')
    plt.ylabel('Kayƒ±p')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    return nn_model, test_accuracy, history

# Neural network modelini eƒüit
print("üß† Neural Network modeli eƒüitiliyor...")
nn_model, nn_accuracy, nn_history = train_neural_network(train_df, test_df, feature_cols)
print("‚úÖ 5. ADIM tamamlandƒ±!")
```
